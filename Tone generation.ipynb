```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf

print('Libraries imported ...')
```

**Parameters**

```python
# Sampling frequency
Fs=8e3
Td=10

print('[=] Sampling frequency : {} Hz'.format(Fs))
print('[=] Signal duration : {} Sec'.format(Td))
```

**Generate Time vector**

```python
Ts=1/Fs
t=np.arange(0,Td,Ts)

print('[=] Number of samples {}'.format(t.size))
```

**Generate Tone**

```python
amp=0.8
freq=500

signal=amp*np.sin(2*np.pi*freq*t)
```

**Plot the signal**

```python
plt.figure(figsize=(10,3))
plot_sample=512
plt.plot(t[0:plot_sample],signal[0:plot_sample],'m')
plt.xlabel('Time(s)',fontsize=14)
plt.ylabel('Amplitude',fontsize=14)
plt.xlim([0,max(t[0:plot_sample])])

plt.show()
```

**Save tone to HDD**

```python
# Make Fs as integer
Fs=int(Fs)

# Convert to 32 bit float as supported by scipy
signal=signal.astype('float32')

# Convert to mono channel
signal=signal.reshape(-1,1)

wf.write('tone.wav',Fs,signal)

print('[=] Tone signal is stored at the current directory')
```
